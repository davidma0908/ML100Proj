# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline
```

> -  寫出 ReLU & dReLU 一階導數

```python
def ReLU(x):
    return abs(x) * (x > 0)

def dReLU(x):
    return 1 * (x > 0)
```

```python
x = np.linspace(-10, 10, 100)

plt.plot(x, ReLU(x))
plt.title('ReLU Function')
plt.grid()
plt.show()
```

```python
plt.plot(x, dReLU(x))
plt.title('dReLU ')
plt.grid()
plt.show()
```
